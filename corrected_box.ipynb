```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import astropy.coordinates as coord
from astropy             import units as u
from mpl_toolkits.mplot3d import Axes3D
%matplotlib inline
```

```python
theta = np.random.uniform(-90,90,100000)
phi = np.random.uniform(-180,180,100000)
```

```python
ra = phi
#ra = ra.wrap_at(180*u.degree)
dec = theta

fig = plt.figure(figsize=(16,12))
#fig.suptitle('Map of cluster centers', fontsize=16)
ax = fig.add_subplot(111,projection="mollweide")
ax.set_title("Randomly distributed points on a sphere\n",fontsize=30)
ax.scatter(ra*2*np.pi/360, dec*2*np.pi/360,s=1);
```

```python
center_ra=10
center_dec=10
theta_max=10
delta_dec=dec-center_dec
delta_ra=np.arccos( np.sin(center_dec*2*np.pi/360)**2 + np.cos(center_dec*2*np.pi/360)**2*np.cos((ra-center_ra)*2*np.pi/360) )
delta_ra2=delta_ra*360/2/np.pi
idx=np.where( (np.abs(delta_dec)<theta_max) & (np.abs(delta_ra2)<theta_max) )
sep=np.arccos( np.sin(center_dec*2*np.pi/360)*np.sin(dec*2*np.pi/360) + np.cos(center_dec*2*np.pi/360)*np.cos(dec*2*np.pi/360)*np.cos((ra-center_ra)*2*np.pi/360) )
```

```python
idx2=np.where( np.abs(sep*360/2/np.pi)<theta_max )
```

```python
ra2 = phi[idx]
dec2 = theta[idx]

ra3=phi[idx2]
dec3=theta[idx2]
fig = plt.figure(figsize=(16,12))
#fig.suptitle('Map of cluster centers', fontsize=16)
ax = fig.add_subplot(111,projection="mollweide")
ax.set_title("Spherical Box of width theta_max\n",fontsize=30)
ax.scatter(ra2*2*np.pi/360, dec2*2*np.pi/360,s=1);
ax.scatter(ra3*2*np.pi/360, dec3*2*np.pi/360,s=1);
```
